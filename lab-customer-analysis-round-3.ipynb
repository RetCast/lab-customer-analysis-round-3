```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('files_for_lab/csv_files/marketing_customer_analysis.csv')
df.head(2)
```

```python
df.info()
```

```python
df.describe().T
```

```python
df.Response.value_counts().plot(kind = 'bar')
plt.title('Number of responses', color='black')
```

```python
sns.countplot(data=df, x="Response", hue="Sales Channel", color='green')
plt.title('Response Rate by the Sales Channel', color='black')
```

```python
sns.countplot(data=df, x="Sales Channel", hue="Response")
plt.title('Response Rate by the Sales Channel', color='black')
```

```python
sns.displot(data=df, x='Total Claim Amount', hue='Response', kind='kde')
plt.title("Response Rate by the Total Claim Amount")
```

```python
sns.displot(data=df, x='Income', hue='Response', kind= 'kde')
plt.title("Response Rate by Income")
```
